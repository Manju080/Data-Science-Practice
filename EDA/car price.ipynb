```python

k=10
corfun=corrmat.nlargest(k,'price')['price'].index
cm=np.corrcoef(ds[corfun].values.T)
sns.heatmap(cm)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=corfun.values, xticklabels=corfun.values)
```
