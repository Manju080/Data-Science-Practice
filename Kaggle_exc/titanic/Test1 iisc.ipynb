```python
data = pd.read_csv('bike_data_new.csv')
def duration_trips_by_manju(data):
    #This function removes all trips of duration 0 minutes and prints the following values on the console:

    start_time = time.time()  #Starting the timer to measure the runtime of the function

    # Remove trips with zero duration
    data = data[data['started_at'] != data['ended_at']]
    #duration of trip
    data['duration'] = (pd.to_datetime(data['ended_at']) - pd.to_datetime(data['started_at'])).astype('timedelta64[ms]')

    # Calculate maximum and minimum trip durations
    max_duration = data['duration'].max()
    min_duration = data['duration'].min()

    # Calculate total number of trips corresponding to the minimum duration
    total_trips_min_duration = len(data[data['duration'] == min_duration])

    # Calculate percentage of total circular trips
    total_circular_trips = len(data[data['start_lat'] == data['end_lat']][data['start_lng'] == data['end_lng']])
    percentage_circular_trips = (total_circular_trips / len(data)) * 100

    end_time = time.time()  # Ending the timer to measure the runtime of the function
    runtime = end_time - start_time

   
    print("Maximum duration of the trip (in minutes):", max_duration)
    
    print("Minimum duration of the trip (in minutes):", min_duration)
   
    print("Total number of trips corresponding to the minimum duration:", total_trips_min_duration)
    
    print("Percentage of total circular trips:", round(percentage_circular_trips, 2), "%")
    
    print("Total runtime for the function (in seconds):", round(runtime, 4))

    return data

# Call the function
data = duration_trips_by_manju(data)
```
